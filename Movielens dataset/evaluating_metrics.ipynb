```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

TfidfVectorizer function from scikit-learn, which transforms text to feature vectors that can be used as input to estimator.

Cosine Similarity to calculate a numeric value that denotes the similarity between two books.

```python
tf=TfidfVectorizer(stop_words='english')

df1['genres']=df1['genres'].fillna('')

#construct the required TF-TD matrix by fitting and transforming the data
tfid_matrix=tf.fit_transform(df1['genres'])

# output the shape of tfidf_matrix
tfid_matrix.shape
```

```python
# import linear_kernel
from sklearn.metrics.pairwise import linear_kernel

# compute the cosine similarity
cosine_sim=linear_kernel(tfid_matrix, tfid_matrix)
```

```python
# construc a reverse map of indices and movie titles

indices=pd.Series(df1.index, index=df1['title']).drop_duplicates()
```

```python
# creating a function to take movie title as input and output as 10 movies similar to the movie
def get_recommendation(title, cosine_sim=cosine_sim):
    idx=indices[title]

    # computing the similarities between movies
    sim_scores=list(enumerate(cosine_sim[idx]))

    # sorting the 10 most similar movies
    sim_scores=sorted(sim_scores, key=lambda x:x[1], reverse=True)

#     get the score of the 10 most similar movies
    sim_scores=sim_scores[1:11]
    print(sim_scores)
        # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return df1['title'].iloc[movie_indices]
```
